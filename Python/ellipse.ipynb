```python
import numpy as np
import matplotlib.pyplot as plt

# パラメーター（楕円の中心座標、長軸、短軸、回転角度）
center_x = 0  # 楕円の中心 x 座標
center_y = 0  # 楕円の中心 y 座標
a = 1        # 長軸の半径
b = 36.5551/90      # 短軸の半径
theta = 139.8828/360*np.pi   # 回転角度（ラジアン）

# 楕円をプロットするための角度（0から2πまで）
t = np.linspace(0, 2 * np.pi, 1000)

# 楕円のパラメーターに基づいてx, y座標を計算
x = center_x + a * np.cos(t) * np.cos(theta) - b * np.sin(t) * np.sin(theta)
y = center_y + a * np.cos(t) * np.sin(theta) + b * np.sin(t) * np.cos(theta)

# プロット
fig, ax = plt.subplots()
ax.plot(x, y, color='blue')

# アスペクト比を等しくし、軸を非表示にする
ax.set_aspect('equal', 'box')
ax.axis('off')

# プロットの表示領域を一定にする（例として -10 から 10 の範囲）
ax.set_xlim(-1.01, 1.01)
ax.set_ylim(-1.01, 1.2)

ax.text(0, 1.05, '宇都宮市(36.5551°,139.8828°)', fontsize=20, ha='center', va='bottom', color='black')

plt.show()
```
